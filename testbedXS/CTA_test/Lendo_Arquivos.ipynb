```python
X_test = pd.DataFrame(df_right.iloc[:10].apply(generate_test_prompt_NextiaJD_few_shot, axis=0), columns=["prediction"])
y_pred = predict(X_test, model, tokenizer)
predictions_right = pd.DataFrame({
                            "Column":df_right.columns, 
                            "Predicted":y_pred
                            })
predictions_right
```

```python
predicted_joinable_columns = []

for i in range(predictions_left.shape[0]):
    column = predictions_left.iloc[i,0]
    predicted_semantic_type = predictions_left.iloc[i,1]
    for j in range(predictions_right.shape[0]):
        if(predicted_semantic_type == predictions_right.iloc[j,1]):
            predicted_joinable_columns.append([column, predictions_right.iloc[j,:].Column, predicted_semantic_type])
```

```python
sameSTColumns = pd.DataFrame(np.array(predicted_joinable_columns), columns=['ColumnLeft','ColumnRight','SemanticType'])
sameSTColumns
```

```python
def jaccard_similarity(set1, set2):
    
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
     
    return intersection / union
```

```python
similarity = []
for i in range(sameSTColumns.shape[0]):
    try:
        similarity.append(jaccard_similarity(set(df_left[sameSTColumns.iloc[i,0]].unique()), set(df_right[sameSTColumns.iloc[i,1]].unique())))
    except:
        similarity.append(jaccard_similarity(set(df_right[sameSTColumns.iloc[i,0]].unique()), set(df_left[sameSTColumns.iloc[i,1]].unique())))
similarity
```

```python
sameSTColumns['JaccardSimilarity'] = similarity
sameSTColumns
```

```python
joinableColumns = sameSTColumns[sameSTColumns['JaccardSimilarity'] > 0.5]
joinableColumns
```
