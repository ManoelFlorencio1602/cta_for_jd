```python
groundTruth = pd.read_csv('groundTruth_testbedXS.csv')
groundTruth = groundTruth[groundTruth['trueQuality'] == 4]
groundTruth = groundTruth.reset_index(drop=True)
groundTruth.head()
```

```python
gt = groundTruth[['ds_name', 'ds_name_2']]
gt = gt.reset_index(drop=True)
gt.head()
```

```python
gt.shape
```

```python
new_gt = pd.DataFrame()
for i in range(gt.shape[0]):
    ds1 = gt.iloc[i,0]
    ds2 = gt.iloc[i,1]

    ds_esq, ds_dir = '', ''
    
    if(ds1 < ds2):
        ds_esq, ds_dir = ds1, ds2
    else:
        ds_esq, ds_dir = ds2, ds1
    
    if(len(new_gt) == 0):
        new_gt = pd.DataFrame({'ds_name': [ds_esq], 'ds_name_2': [ds_dir]})
    else:
        new_gt = pd.concat([new_gt, pd.DataFrame({'ds_name': [ds_esq], 'ds_name_2': [ds_dir]})], axis=0)

new_gt = new_gt.drop_duplicates()
new_gt = new_gt.reset_index(drop=True)
print(new_gt.shape)
```
