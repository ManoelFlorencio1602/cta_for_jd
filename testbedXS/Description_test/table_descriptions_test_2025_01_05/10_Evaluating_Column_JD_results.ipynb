```python
def generate_keys(df):
    keys = []
    for i in range(df.shape[0]):
        key = df.iloc[i, 0] + '#' + df.iloc[i, 1] + '#' + df.iloc[i, 2] + '#' + df.iloc[i, 3]
        keys.append(key)
    return keys
```

```python
def sorting_tablenames(df):
    sorted_df = pd.DataFrame()
    for i in range(df.shape[0]):
        ds1 = df.iloc[i,0]
        ds2 = df.iloc[i,1]
        ds3 = df.iloc[i,2] 
        ds4 = df.iloc[i,3]
    
        ds_esq, ds_dir = '', ''
        
        if(ds1 < ds3):
            ds_esq, cl_esq, ds_dir, cl_dir = ds1+'.csv', ds2, ds3+'.csv', ds4
        else:
            ds_esq, cl_esq, ds_dir, cl_dir = ds3+'.csv', ds4, ds1+'.csv', ds2
        
        if(len(sorted_df) == 0):
            sorted_df = pd.DataFrame({'LEFT_TABLE': [ds_esq], 'LEFT_COLUMN': [cl_esq], 'RIGHT_TABLE': [ds_dir], 'RIGHT_COLUMN': [cl_dir]})
        else:
            sorted_df = pd.concat([sorted_df, pd.DataFrame({'LEFT_TABLE': [ds_esq], 'LEFT_COLUMN': [cl_esq], 'RIGHT_TABLE': [ds_dir], 'RIGHT_COLUMN': [cl_dir]})], axis=0)

    sorted_df = sorted_df.drop_duplicates()
    sorted_df = sorted_df.reset_index(drop=True)

    return sorted_df
```

```python
groundTruth = pd.read_csv('Description_test/table_descriptions_test_2025_01_05/groundTruth_df_20241104.csv')
groundTruth.head()
```

```python
jd_df = pd.read_csv('Description_test/table_descriptions_test_2025_01_05/column_jd.csv')
jd_df = jd_df[jd_df['COLUMN_JOINABLE'] == 'Yes']
jd_df = jd_df[['LEFT_TABLE','LEFT_COLUMN','RIGHT_TABLE','RIGHT_COLUMN']]
jd_df = sorting_tablenames(jd_df)
jd_df.head()
```
