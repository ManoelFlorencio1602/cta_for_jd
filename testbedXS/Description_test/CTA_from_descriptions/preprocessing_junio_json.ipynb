```python
import json
import pandas as pd
import os
```

```python
with open('nextiajdxs_description_junio.json', 'r') as file:
    d = json.load(file)
```

```python
d
```

```python
df = pd.DataFrame({}, columns=['TableName','Column','Description'])

for tb_name in d.keys():
    for column_name in d[tb_name]['attributes'].keys():
        s = pd.DataFrame({'TableName': [tb_name], 'Column':[column_name], 'Description':[d[tb_name]['attributes'][column_name]]})
        if(len(df) == 0):
            df = s
        else:
            df = pd.concat([df, s], axis=0)

df.head()
```

```python
df.to_csv('nextiajdxs_description_junio.csv',index=False)
```
