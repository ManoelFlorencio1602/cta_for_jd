```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
ground_truth = pd.read_csv('opendata_join_ground_truth.csv')
ground_truth.head()
```

```python
opendata_sg_ground_truth = ground_truth[(ground_truth.query_table.str.contains("SG_CSV")) & (ground_truth.candidate_table.str.contains("SG_CSV"))]
opendata_sg_ground_truth.head()
```

```python
opendata_sg_ground_truth.shape
```

```python
for i in range(opendata_sg_ground_truth.shape[0]):
  print(opendata_sg_ground_truth.iloc[i,0])
```

```python
def clean_table_name(table_name):
    if("__" in table_name):
        return table_name.split("__")[0] + ".csv"
    else:
        return table_name
```

```python
opendata_sg_ground_truth.query_table = opendata_sg_ground_truth.query_table.apply(clean_table_name)
opendata_sg_ground_truth.candidate_table = opendata_sg_ground_truth.candidate_table.apply(clean_table_name)
```

```python
opendata_sg_ground_truth.head()
```

```python
def removing_same_tables(df):
    df_clean = pd.DataFrame(columns=['query_table','candidate_table','query_column','candidate_column'])
    for i in range(df.shape[0]):
        if(df.iloc[i,0] != df.iloc[i,1]):
            df_clean = pd.concat([df_clean, pd.DataFrame([df.iloc[i,:]], columns=df_clean.columns)])
    return df_clean
```

```python
opendata_sg_ground_truth_clean = removing_same_tables(opendata_sg_ground_truth)
opendata_sg_ground_truth_clean.head()
```

```python
opendata_sg_ground_truth_clean.shape
```

```python
def get_table_id(table_name):
    return int(table_name.split(".")[0].split("CSV")[1])
```

```python
def sort_table_names(df):
    df_copy = df.copy()
    for i in range(df.shape[0]):
        if(get_table_id(df_copy.iloc[i,0]) > get_table_id(df_copy.iloc[i,1])):
            df_copy.iloc[i,0], df_copy.iloc[i,1] = df_copy.iloc[i,1], df_copy.iloc[i,0]
    return df_copy
```

```python
opendata_sg_ground_truth_clean_sorted = sort_table_names(opendata_sg_ground_truth_clean)
opendata_sg_ground_truth_clean_sorted
```

```python
opendata_sg_ground_truth_clean_sorted_unique = opendata_sg_ground_truth_clean_sorted.drop_duplicates(keep='first')
opendata_sg_ground_truth_clean_sorted_unique
```

```python
opendata_sg_ground_truth_clean_sorted_unique.shape
```

```python
opendata_sg_ground_truth_clean_sorted_unique.to_csv('opendata_join_ground_truth_clean.csv', index=False)
```
