```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matpotlib inline
```

```python
df = pd.read_csv('opendata_join_ground_truth.csv')
df.head()
```

```python
def clean_table_name(table_name):
    if("__" in table_name):
        return table_name.split("__")[0] + ".csv"
    else:
        return table_name
```

```python
df.query_table = df.query_table.apply(clean_table_name)
df.candidate_table = df.candidate_table.apply(clean_table_name)
```

```python
df.head()
```

```python
def removing_same_tables(df):
    df_clean = pd.DataFrame(columns=['query_table','candidate_table','query_column','candidate_column'])
    for i in range(df.shape[0]):
        if(df.iloc[i,0] != df.iloc[i,1]):
            df_clean = pd.concat([df_clean, pd.DataFrame([df.iloc[i,:]], columns=df_clean.columns)])
    return df_clean
```

```python
df = removing_same_tables(df)
df.head()
```
